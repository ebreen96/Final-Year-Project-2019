## Packages used

```python
df_import.head(15).plot(kind = 'bar', x = 'imports', figsize = (12,6), fontsize = 12)
plt.xlabel('Packages', fontsize=16)
plt.ylabel('Number of times used', fontsize=16)
plt.grid(True)
plt.title('Bar Chart showing the most used packages', fontsize=18)
plt.savefig('../data/Images/imports.png', bbox_inches='tight')
```

Above, we can see numpy is the most used library with 109,356 occurrences thus appearing in over 50\% of the notebooks. The second most used library is matplotlib and the third pandas, indicating a lot of users are creating visualizations in their notebooks. There were 23,727 packages appearing in less than 100 notebooks and 11,362 packages making only one appearance. This is an indication that there are a small amount of packages that are used regularly by notebook users.

## Heading Levels

```python
df_header['Heading_Level'].value_counts().sort_index(ascending=False).plot(kind='bar', figsize = (12,6), 
                                                                           fontsize = 12, 
                                                                           color = 'blue')
plt.xlabel('Heading level', fontsize=16)
plt.ylabel('Number of times used', fontsize=16)
plt.grid(True)
plt.title('Bar Chart showing use of headings', fontsize=18)
plt.savefig('../data/Images/headers.png', bbox_inches='tight')
```

In Jupyter notebooks, headers or titles are created by using the hashtag symbol (\#) in a markdown cell. The levels of the headers (ie. the size of headings and sub-headings) are denoted by the number of hashtags used. \# level 1 header is the largest, '\#\#' level 2 is the second largest and so on. A bar chart can be seen showing which header level is used most often. The level 3 header is the most popular followed by level 2 then by 1. This is a surprising result but can be validated. When you format notebooks, a level 1 header would be used for introductory titles and level 2 or 3 would be used for sub-headings/sub-titles which there would be more of. Altogether, there were 788,836 headers used which is a staggering amount considering there were just 200,000 notebooks. This use of headings and sub-headings in notebooks shows that many users are focusing on formatting and the presentation of their notebooks. Thus, showing us that most users are approaching their notebooks with the expectation of other people reading or interpreting it. This again is promising for the future of the literate programming style of writing and documenting code.

## Visualisation use

```python
my_colours = 'bgrymc'
df_display.loc[1].sort_values(ascending = False).plot.bar(figsize = (12, 6), fontsize = 14,
                                                         color = my_colours)
plt.xlabel('Type of plot', fontsize=16)
plt.ylabel('Number of times used', fontsize=16)
plt.grid(True)
plt.title('Bar Chart showing types of plots used in notebooks', fontsize=18)
plt.savefig('../data/Images/visualisations.png', bbox_inches='tight')
```

Histograms, bar charts, line plots, scatter plots and pie charts corresponds to 164,972 of the 363,720 instances of display data in the notebooks. Line plots are the most popular graphs appearing 95,687 times in the notebooks. As line graphs are the default when using plot(), this result was not surprising. Predictably, pie charts were the least common appearing only 1051 times. Interestingly, bar charts appeared 4th most popular with just 12,958 appearances. From this analysis it was also found that notebook users favoured using type .bar() or .hist() over specifying the graph using kind = bar or kind = hist.
